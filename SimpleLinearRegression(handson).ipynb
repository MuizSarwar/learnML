```python
#import all libraies:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#creat a dataframe:
data = {
    "Experience_Years": [0, 1, 2, 3, 4, 5, 6, 7],
    "Salary": [13000, 19000, 23000, 26000, 31000, 34000, 40000, 42000]
}

df = pd.DataFrame(data)
df
```

```python
#creat and fit model :
lrModel = LinearRegression()
lrModel.fit(df[["Experience_Years"]], df["Salary"])
```

```python
#vizualization of actual data ( using scatter plot )
plt.scatter(df["Experience_Years"], df["Salary"],color='blue',label="Actual Data")

#vizualization of predicted data ( using plot )
predicted_salary = lrModel.predict(df[["Experience_Years"]])
plt.plot(df["Experience_Years"], predicted_salary,color='red',label="Predicted Data")

plt.legend()
plt.xlabel("Experience_Years")
plt.ylabel("Salary")
plt.title("Linear Regression Model")
plt.grid(True)
plt.show()
```

```python
#let's predict a value for Experience year = 4.2 , 5.2 , 5.9 years
pred_val = lrModel.predict([[4.2],[5.2],[5.9]])
pred_val
```

```python
#let's vizualized the final pred_val ( using scatter plot )

#vizualization of actual data ( using scatter plot )
plt.scatter(df["Experience_Years"], df["Salary"],color='blue',label="Actual Data")

#vizualization of predicted data ( using plot )
predicted_salary = lrModel.predict(df[["Experience_Years"]])
plt.plot(df["Experience_Years"], predicted_salary,color='red',label="Predicted Data")

#vizualization of pred_val ( using scatter plot )
plt.scatter([[4.2],[5.2],[5.9]],pred_val,color='green',label="pred_val")

plt.legend()
plt.xlabel("Experience_Years")
plt.ylabel("Salary")
plt.title("Linear Regression Model")
plt.grid(True)
plt.show()
```
