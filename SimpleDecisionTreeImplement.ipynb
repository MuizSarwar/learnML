```python
#import all libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,f1_score,roc_curve,auc
```

```python
# Sample synthetic dataset with 8 entries
data = {
    "weather": ["Sunny", "Sunny", "Rainy", "Sunny", "Rainy", "Sunny", "Sunny", "Rainy"],
    "windy": [0, 0, 1, 1, 0, 1, 0, 1],  # Binary values for windy
    "play": [1, 1, 0, 1, 0, 1, 1, 0]    # Binary values for play
}

# Creating the DataFrame
df = pd.DataFrame(data)
df
```

```python
#encoding categorical feature "weather" using map
df["weather_num"] = df["weather"].map({"Sunny":0,"Rainy":1})
df
```

```python
#Define X and Y values for the dataset:
X = df[["weather_num","windy"]]
Y = df["play"]
```

```python
#create a model using DecisionTreeClassifier and train the model
tree = DecisionTreeClassifier(max_depth=3, random_state=42)
tree.fit(X,Y)
```

```python
#visualize the created model :
plt.figure(figsize=(8,6))
plot_tree(
    tree,
    feature_names=X.columns,
    class_names=["No","Yes"],
    filled=True,
    rounded=True
)

plt.show()
```

```python
#lets  predicts a value:
example = pd.DataFrame([[1,0]],columns=["weather_num","windy"])
predicted_value = tree.predict(example)[0]
predicted_value
```
